## ต่อแกน X

```python
personal_data = data_df[['No.','sex','age','nationality']]
personal_data.head()
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1) # .sample(frac=1) สุ่มสลับข้อมูล
covid_data.head()
```

เอาตารางมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data)
personalcovid_data.head()
```

```python
personalcovid_data = covid_data.merge(personal_data)
personalcovid_data.head()
```
