## ต่อแกน X

```python
personal_data = data_df[['No.','sex','age','nationality']]
personal_data.head()
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1) # .sample(frac=1) สุ่มสลับข้อมูล
covid_data.head()
```

เอาตารางมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data)
personalcovid_data.head()
```

```python
personalcovid_data = covid_data.merge(personal_data)
personalcovid_data.head()
```

## List(Row Oriented)

```python
data = [['Jones LLC',150,200,140],
            ['Alpha',200,210,215],
            ['Blue',50,90,95]]
headers = ['account','Jan','Feb','Mar']
```

```python
df2 = pd.DataFrame.from_records(data,columns=headers)
df2
```

## Dictionary(Column Oriented)

```python
data_dict = {'account' : ['Jones LLC','Alpha Co','Blue Inc'],
             'Jan' : [150,200,50],
             'Feb' : [200,210,90],
             'Mar' : [140,215,95]}
```
