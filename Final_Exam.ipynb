## รวมตารางข้อมูล alldb และ hospital เข้าด้วยกัน โดยใช้การ merge

```python
data = drop_alldb.merge(drop_hospital,how='left',left_on='site',right_on='hcode')
data
```

```python
data.isnull().sum() # เช็คค่า missing
```

```python
drop_data = data.dropna() # จัดการค่า missing โดยการ drop
drop_data
```
